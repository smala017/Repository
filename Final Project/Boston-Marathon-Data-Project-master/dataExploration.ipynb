```python
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2021, 2022]
times = []

for i in range(len(years)):
    print(years[i])
    df = pd.read_csv(f"data/results{years[i]}.csv", encoding="ISO-8859-1")
    print(len(df["seconds"]))
    times.append(df["seconds"])

plt.boxplot(times, labels=years, vert=False)
```

```python
result = pd.DataFrame()
i=0

for yearResultsFull in resultsDFs:
    yearResultsSmall = pd.DataFrame()
    year = int(years[i])
    print(year)

    yearResultsSmall["year"] = 9999 # temporary just to get it at the front
    yearResultsSmall["gender"] = yearResultsFull["gender"]
    yearResultsSmall["official_time"] = yearResultsFull["official_time"]
    yearResultsSmall["seconds"] = yearResultsFull["seconds"]
    yearResultsSmall["year"] = float("-" + str(year) + ".0")

    if i == 0:
        result = yearResultsSmall
    else:
        result = pd.concat([result, yearResultsSmall], ignore_index=True)

    i+=1

result = result[((result["gender"] == "M") | (result["gender"] =="F"))].reset_index(drop=True)
result.to_csv('data/condensedResults.csv', index=False)


```
